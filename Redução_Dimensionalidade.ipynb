```python
import matplotlib.pyplot as plt
from google.colab import drive
import matplotlib.image as mpimg
import numpy as np
drive.mount('/content/drive')
```

```python
def mostrar_imagem(imagem, titulo, cmap=None):
    plt.figure(figsize=(6, 6))
    plt.imshow(imagem, cmap=cmap)
    plt.title(titulo)
    plt.axis('off')
    plt.show()
```

```python
imagem_rgb = mpimg.imread('drive/MyDrive/rick.jpg')
mostrar_imagem(imagem_rgb, 'Imagem Colorida',)
```

```python
r, g, b = imagem_rgb[:, :, 0], imagem_rgb[:, :, 1], imagem_rgb[:, :, 2]
imagem_cinza = 0.2989 * r + 0.5870 * g + 0.1140 * b
mostrar_imagem(imagem_cinza, 'Imagem em Escala de Cinza', cmap='gray')
```

```python
limiar = 128
imagem_binarizada = np.where(imagem_cinza < limiar, 0, 255)
mostrar_imagem(imagem_binarizada, 'Imagem Binarizada', cmap='gray')
```
